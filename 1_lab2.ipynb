# Multi-Model Physics Question Generator for IIT JEE

## Overview
This Python script demonstrates generating and solving IIT JEE Physics questions using multiple AI models including OpenAI, DeepSeek, and optionally Gemini.

## Functionality

### 1. **Environment Setup**
- Loads API keys from `.env` file using `dotenv`
- Supports multiple AI providers: OpenAI, DeepSeek, and Gemini
- Uses `override=True` to ensure latest environment variables are loaded

### 2. **Question Generation (OpenAI)**
- Sends a request to GPT-4o-mini: *"Please come up with a challenging Physics question for IIT JEE exam. Only Question no answer"*
- Stores the generated physics problem for further processing

### 3. **Model Exploration**
- Lists all available models from DeepSeek API
- Displays model IDs (e.g., `deepseek-chat`, `deepseek-reasoner`)

### 4. **Multi-Model Comparison**
- Sends the same generated question to different AI models:
  - **OpenAI GPT-4o-mini** (first response)
  - **DeepSeek-Chat** (via OpenAI-compatible API)
  - **DeepSeek-Reasoner** (optional)
  - **Gemini-Pro** (optional, via google-generativeai)
- Collects and stores responses from each model

### 5. **Output Display**
- Uses IPython's `Markdown` for formatted display of responses
- Tracks competitors and their answers in lists

## Key Features
- **Cross-Platform**: Works with multiple AI providers using similar APIs
- **Modular**: Easy to add/remove models
- **Educational**: Demonstrates how different AI models approach physics problems
- **Configurable**: API keys managed via environment variables

## Dependencies
- `openai`: For OpenAI and DeepSeek APIs
- `python-dotenv`: For environment variable management
- `IPython`: For markdown display
- `google-generativeai`: Optional, for Gemini integration


## Use Cases
- Comparing responses from different AI models
- Generating practice questions for IIT JEE preparation
- Testing API integrations with multiple providers
- Educational demonstrations of AI capabilities

```python
import os
import json
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
load_dotenv(override=True)
open_api_key  = os.getenv("OPENAI_API_KEY")
deep_seek_api_key = os.getenv("DEEP_SEEK_API_KEY")

request ="Please come up with a challenging Physics question for IIT JEE exam. Only Question no answer"
message  = [{"role":"user", "content": request}]
openai = OpenAI()
response = openai.chat.completions.create(
    model="gpt-4o-mini",
    messages=message
)
question=response.choices[0].message.content
print(question)
compititor = [];
answer = [];
message =  [{"role":"user", "content": question}]


deepseekclient = OpenAI(api_key=deep_seek_api_key, base_url="https://api.deepseek.com/v1")
models = deepseekclient.models.list()
for model in models:
    print(model.id)

model_name ="gpt-4o-mini"
response = openai.chat.completions.create(
    model=model_name,
    messages=message
)
answer = response.choices[0].message.content
display(Markdown(answer))
compititor.append(model_name)
compititor.append(answer)

from  google import genai 
gemini_api_key = os.getenv("GEMINI_API_KEY") # Note: Usually named GOOGLE_API_KEY for this SDK
model_name= "gemini-3-flash-preview"
#googleclient = OpenAI(api_key=gemini_api_key, base_url="https://generativelanguage.googleapis.com/v1beta/openai/")
googleclient = genai.Client(api_key=gemini_api_key)
response = googleclient.models.generate_content(
    model=model_name,
    contents=question
)
answer= response.choices[0].message.content
display(Markdown(answer))
compititor.append(model_name)
compititor.append(answer)

ollama = OpenAI(base_url="http://localhost:11434/v1", api_key="ollama")
model_name = "llama3.1"
response = ollama.chat.completions.create(
    model=model_name,
    messages=message
)
answer = response.choices[0].message.content
display(Markdown(answer))
compititor.append(model_name)
compititor.append(answer)
```
